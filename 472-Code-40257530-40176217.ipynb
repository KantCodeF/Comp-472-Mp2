```python
#rush hour game methods definition
import numpy as np
import time
import pandas as pd
#based on gameboard and car letter return x or y
def moveAxis(gameboard,car):
    for char in range (len(gameboard)):
        
        if gameboard[char]==car and 'p1' not in locals():
            p1=char
        elif gameboard[char]==car:  
            p2=char
            break
    if (p1-p2)==-6:
        return 'y'
    if (p1-p2)==-1:
        return 'x'

    
#read text file and return  6x6 array of gameboard and dict of fuel level 
def readfile(filename):
    gameboards=[]
    carFuels=[]
    f = open(filename, "rt")
    count=0
    for x in f:
        gameboard=[]
        carFuel={}
        if x[0]=="#":
            continue
        if len(x)==1:
            continue
        else:
            x=x.strip("\n")
            token=x.split()
            for i in range(len(token)):        
                if i==0:
                    count+=1
                    for j in range(36):
                        gameboard.insert(len(gameboard),token[i][j])
                        if token[i][j] not in  carFuel and  token[i][j] !='.':
                            carFuel[token[i][j]]=100
                elif i>=1: 
                    
                    car=token[i][0]
                    fuelLevel=int(token[i][1])
                    carFuel.update({car:fuelLevel})
        gameboard=np.array(gameboard).reshape(6,6)
        gameboards.insert(count-1,gameboard)
        carFuels.insert(count-1,carFuel)        
                          
    f.close()
    return gameboards,carFuels

#adds the valet service to the game
def valetService(gameboard,car):
    valetBoard=gameboard.copy()
    if car=="A":
        return valetBoard
    if valetBoard[17]==car:
        valetBoard[:] = [x if x != car else "." for x in valetBoard]
    return valetBoard    
    
#makes a dict of possible moves
def possiblemoves(gameboard,fuelLevel):
    gameboard=gameboard.flatten().tolist()
    moves={}
    
    for car in fuelLevel.keys():
        if car in gameboard:
            axis=moveAxis(gameboard,car)
            carMoves={}   
            if(axis=='x'):       
                left=MovePossibleLeft(gameboard,car)
                right=MovePossibleRight(gameboard,car)       
                if(left!=0 and fuelLevel[car]>left):
                    carMoves.update({"left":left})
                if(right!=0 and fuelLevel[car]>right):
                    carMoves.update({"right":right})

            if(axis=='y'):
                up=MovePossibleUp(gameboard,car)
                down=MovePossibleDown(gameboard,car)
                if(up!=0 and fuelLevel[car]>up):

                    carMoves.update({"up":up})
                if(down!=0 and fuelLevel[car]>down):
                    carMoves.update({"down":down})

            if(len(carMoves)>0):
                moves.update({car:carMoves})
    return moves

#return spaces car can move left
def MovePossibleLeft(gameboard,car):
    count=0
    index=gameboard.index(car)
    while(index%6 != 0):
        index-=1
        if(gameboard[index]=="."):
            count+=1
            continue
        else:
            break
    return count


#return spaces car can move right
def MovePossibleRight(gameboard,car):
    count=0
    index=gameboard.index(car)
    while(index%6 != 5):
        index+=1
        if(gameboard[index]=="."):
            count+=1
            continue
        if(gameboard[index]==car):
            continue
        else:
            break
    return count


#return spaces car can move Up
def MovePossibleUp(gameboard,car):
    count=0
    index=gameboard.index(car)-6
    while(index>=0):
        
        if(gameboard[index]=="."):
            count+=1
            index-=6
            continue
        else:
            break
    return count

#return spaces car can move down
def MovePossibleDown(gameboard,car):
    count=0
    index=gameboard.index(car)+6
    
    while(index<=35):
        if(gameboard[index]=="."):
            count+=1
            index+=6
            continue
        if(gameboard[index]==car):
            index+=6
            continue
        else:
            break
    return count
#return possible gameboard with possible moves based on a 6x6 gameboard and fuel dict
def possibleboard(gameboard, fuel):
    possiblemove=possiblemoves(gameboard,fuel)
    gameboards=[]
    
    gameboard=gameboard.flatten().tolist()
    
    for car in possiblemove:
        
        for direction in possiblemove[car]:
            count=1
            newBoard=gameboard.copy()
            newFuel=fuel.copy()
            index=newBoard.index(car)
            
            
            if direction=="up":
                while possiblemove[car][direction]>=count:
                    if newBoard[index-6]==".":
                        newBoard[index-6]=car
                        i=1

                        while (newBoard[index+(i*6)]==car)  :
                            i+=1
                            if (index+i*6>35):
                                break
                        newBoard[index+(i*6)-6]='.'
                        option=np.array(newBoard).reshape(6,6)
                        newFuel[car]=newFuel[car]-1
                        gameboards.append((option,newFuel.copy(),direction,count))
                        count+=1
                        index-=6
                        continue
            if direction=="down":
                while possiblemove[car][direction]>=count:
                    
                    newBoard[index]='.'
                    i=1
                    while(newBoard[index+(i*6)]==car) :
                        i+=1
                        if (index+i*6>35):
                            break
                    newBoard[index+(i*6)]=car
                    option=np.array(newBoard).reshape(6,6)
                    newFuel[car]=newFuel[car]-1
                    gameboards.append((option,newFuel.copy(),direction,count))
                    count+=1
                    index+=6
                    continue
            if direction== "left":
                while possiblemove[car][direction]>=count: 
                    newBoard[index-1]=car
                    i=1
                    while(newBoard[index+i]==car) :
                        i+=1
                        if ((index+i)%6)==0:
                            break
                    newBoard[index+i-1]='.'
                    option=np.array(newBoard).reshape(6,6)
                    newFuel[car]=newFuel[car]-1
                    gameboards.append((option,newFuel.copy(),direction,count))
                    count+=1
                    index-=1
                    continue
            if direction=="right":
                while possiblemove[car][direction]>=count:
                    newBoard[index]='.'
                    i=1
                    while (newBoard[index+i]==car) :
                        i+=1
                        if ((index+i)%6)==5:
                            break
                    newBoard[index+i]=car
                    newBoard=valetService(newBoard,car)
                    option=np.array(newBoard).reshape(6,6)
                    newFuel[car]=newFuel[car]-1
                    gameboards.append((option,newFuel.copy(),direction,count))
                    count+=1
                    index+=1
                    continue
    return gameboards
            
```

```python
class Tree:

    def __init__(self,g,gameboard,fuel,direction,count,parent,h,f):
        self.g = g
        self.gameboard=gameboard
        self.fuel=fuel
        self.direction=direction
        self.count=count
        self.nodes = []
        self.parent=parent
        self.f = f
        self.h = h

    def add_node(self, g,gameboard,fuel,direction,count,parent,h,f):
        self.nodes.append(Tree(g,gameboard,fuel,direction,count,parent,h,f))

    def __repr__(self):
        return f"Tree({self.g,self.gameboard,self.fuel,self.direction,self.count}): {self.nodes}"


```

```python
def isGoal(gameboard):
    if gameboard[17]=="A":
        return True
    else:
        return False

#number of blocking veheicles
def h1(gameboard):
    if(isGoal(gameboard)):
        return 0
    else: 
        A_position = gameboard.index("A")
        blocked_cars = set()
        for i in range(A_position, 17):
            if gameboard[i] == "A":
                continue
            elif gameboard[i] == ".":
                continue
            elif gameboard[i] != "A":
                blocked_cars.add(gameboard[i])
        return len(blocked_cars)

def h2(gameboard):
    if(isGoal(gameboard)):
        return 0
    else: 
        A_position = gameboard.index("A")
        blocked_position_counter = 0
        for i in range(A_position, 17):
            if gameboard[i] == "A":
                continue
            elif gameboard[i] == ".":
                continue
            elif gameboard[i] != "A":
                blocked_position_counter = blocked_position_counter + 1
        return blocked_position_counter
    
# Number of The value of h1 multiplied by a constant λ of your choice, where λ = 5
def h3(gameboard):
    return h1(gameboard) * 5

# Number of The value of h2 multiplied by a constant λ of your choice, where λ = 2
def h4(gameboard):
    return h2(gameboard) * 2
```

```python
#h(n) (estimate of lowest cost from node to goal)
#queue sorted by h(n)
def take6(elem):
    return elem[6]

def GBFS(gameboard,fuel,h):
    openq = [(0,gameboard,fuel,None,None,None,0,0)]
    visited = []
    visited_node = []
    tree = Tree(None,None,None,None,None,None,None,None)
    statesExplored=0

    while openq:
        openq.sort(key=take6)
        next_item = openq.pop(0)
        statesExplored = statesExplored + 1
        item_board = next_item[1].flatten().tolist()
        item_fuel = next_item[2]
        tree.add_node(next_item[0],next_item[1],next_item[2],next_item[3],next_item[4],next_item[5],next_item[6],next_item[7])
        tree = tree.nodes[-1] #traverse to the next child
        if item_board in visited:
            continue
        visited.append(item_board)
        visited_node.append(tree)
        if(isGoal(item_board)):
            #change to output format file later
            #Output solution path
            return tree,statesExplored,visited_node
        else:            
            possibleboards = possibleboard(next_item[1],item_fuel);
            for possiblemoves in possibleboards:
                next_board = possiblemoves[0]
                next_fuel = possiblemoves[1]
                direction = possiblemoves[2]
                count = possiblemoves[3]
                h_score = h(next_board.flatten().tolist())
                tree.add_node(0,next_board,next_fuel,direction,count,tree,h_score,h_score)
                openq.append((0,next_board,next_fuel,direction, count,tree,h_score,h_score))
    return "No solution",statesExplored,visited_node

def write_sol_files(gameboards, fuel, algo, algo_string, h, h_string):
    for i in range (0,len(gameboards)):
        #write solution into solution file
        if h != None:
            start = time.time()
            goalstate=algo(gameboards[i],fuel[i], h)
            end = time.time()
            f = open(algo_string + "-"+ h_string + "-sol-" + str(i+1) + ".txt", "w")
        else:
            start = time.time()
            goalstate=algo(gameboards[i],fuel[i])
            end = time.time()
            f = open(algo_string + "-sol-" + str(i+1) + ".txt", "w")
        visited_node = goalstate[2]
        duration = "{:.2f}".format(end - start)
        f.write("--------------------------------------------------------------------------------\n\n")
        initial_config = ''.join(gameboards[i].flatten().tolist())
        f.write("Initial board configuration: " + initial_config + "\n\n")
        for line in gameboards[i].tolist():
            f.write(''.join(line) + "\n")

        fuelString = []
        for key,value in fuel[i].items():
            fuelString.append(str(key) + ":" + str(value))
        fuelString = ", ".join(fuelString)
        f.write("\nCar fuel available: " + fuelString + "\n")        


        if(isinstance(goalstate[0], Tree) == False):
            f.write("\nSorry, could not solve the puzzle as specified.\n")
            f.write("Error: no solution found\n")
            f.write("\nRuntime: " + str(duration) + " seconds\n")
            f.write("--------------------------------------------------------------------------------")
        else:
            f.write("\nRuntime: " + str(duration) + " seconds\n")   
            f.write("Search path length: " + str(goalstate[1]) + " states\n")


            tree = goalstate[0]
            nu_move = 0
            lines = []
            solution_path = []
            while tree.parent is not None:
                    line = ""
                    current_node_fuel = {key:val for key, val in tree.fuel.items() if val != 100}
                    parent_node_fuel = {key:val for key, val in tree.parent.fuel.items() if val != 100}
                    current_fuel_set = set(current_node_fuel.items())
                    parrent_fuel_set = set(parent_node_fuel.items())
                    diff_item = (current_fuel_set - parrent_fuel_set).pop()
                    car_moved = diff_item[0]
                    car_move_left = diff_item[1]
                    nu_move = nu_move + 1

                    line = line + car_moved
                    space = " "
                    if tree.direction == "down" or tree.direction == "left":
                        space = space + " "
                    elif tree.direction == "up":
                        space = space + "   "
                    line = line + space + tree.direction + " " + str(tree.count) + "       " + str(car_move_left)
                    config = ''.join(tree.gameboard.flatten().tolist())
                    current_fuel_string = []
                    for key,value in current_node_fuel.items():
                        current_fuel_string.append(str(key) + str(value))
                    line = line + " " + config + " " + " ".join(current_fuel_string)
                    lines.append(line)
                    solution_path.append(str(car_moved) + " " + str(tree.direction) + " " + str(tree.count))
                    tree = tree.parent



            f.write("Solution path length: " + str(nu_move) + " moves\n")
            solution_path = "; ".join(reversed(solution_path))
            f.write("Solution path: " + solution_path + "\n\n")
            for l in reversed(lines):
                f.write(l + "\n")

            #write final configuration
            for line in goalstate[0].gameboard.tolist():
                f.write("\n" +''.join(line))
            f.write("\n--------------------------------------------------------------------------------")
            f.close()
            
            #write search file 
            write_search_file(visited_node, algo_string, i, h_string)
            
            

        
```

```python

#h(n) (estimate of lowest cost from node to goal)
#queue sorted by h(n)
#A search
def take7(elem):
    return elem[7]

def A(gameboard,fuel,h):
    openq = [(0,gameboard,fuel,None,None,None,0,0)]
    visited = []
    visited_node = []
    tree = Tree(None,None,None,None,None,None,None,None)
    statesExplored=0

    while openq:
        openq.sort(key=take7)
        next_item = openq.pop(0)
        statesExplored = statesExplored + 1
        item_board = next_item[1].flatten().tolist()
        item_fuel = next_item[2]
        item_g = next_item[0]
        tree.add_node(next_item[0],next_item[1],next_item[2],next_item[3],next_item[4],next_item[5],next_item[6],next_item[7])
        tree = tree.nodes[-1] #traverse to the next child
        
        if item_board in visited:
            continue
        visited.append(item_board)
        visited_node.append(tree)
        if(isGoal(item_board)):
            #change to output format file later
            #Output solution path
            return tree,statesExplored,visited_node
        else:            
            possibleboards = possibleboard(next_item[1],item_fuel);
            for possiblemoves in possibleboards:
                next_board = possiblemoves[0]
                next_fuel = possiblemoves[1]
                direction = possiblemoves[2]
                count = possiblemoves[3]
                h_score = h(next_board.flatten().tolist())
                g_score = item_g + 1
                f_score = h_score + g_score
                tree.add_node(g_score,next_board,next_fuel,direction,count,tree,h_score,f_score)
                openq.append((g_score,next_board,next_fuel,direction, count,tree,h_score,f_score))
    return "No solution",statesExplored,visited_node
```

```python
#f(n), g(n) and h(n)
def write_search_file(visited_node, algo_string, i, h_string):
    if h_string != None:
        f = open(algo_string + "-"+ h_string + "-search-" + str(i+1) + ".txt", "w")
    else:
        f = open(algo_string + "-search-" + str(i+1) + ".txt", "w")
    for node in visited_node:
        f_score = str(node.f)
        h_score = str(node.h)
        g_score = str(node.g)
        config = ''.join(node.gameboard.flatten().tolist())
        node_string = f_score + " " + g_score + " " + h_score + " " + config
        
        f.write(node_string + "\n")
    f.close()
    
    
```

```python
def excel_analysis(gameboards, fuel):
    df = pd.DataFrame(columns=['Puzzle Number','Algorithm','Heuristic','Length of the Solution','Length of the Search Path','Execution Time (in seconds)'])
    for i in range (0,len(gameboards)):
        gameboard = gameboards[i]
        result_UCS = excel_input(gameboard,fuel[i], UCS, 'UCS', None, 'NA', i+1)
        df.loc[len(df)] = result_UCS
        result_GBF1 = excel_input(gameboard,fuel[i], GBFS, 'GBFS', h1, 'h1', i+1)
        df.loc[len(df)] = result_GBF1
        result_GBF2 = excel_input(gameboard,fuel[i], GBFS, 'GBFS', h2, 'h2', i+1)
        df.loc[len(df)] = result_GBF2
        result_GBF3 = excel_input(gameboard,fuel[i], GBFS, 'GBFS', h3, 'h3', i+1)
        df.loc[len(df)] = result_GBF3
        result_GBF4 = excel_input(gameboard,fuel[i], GBFS, 'GBFS', h4, 'h4', i+1)
        df.loc[len(df)] = result_GBF4
        result_A1 = excel_input(gameboard,fuel[i], A, 'A', h1, 'h1', i+1)
        df.loc[len(df)] = result_A1
        result_A2 = excel_input(gameboard,fuel[i], A, 'A', h2, 'h2', i+1)
        df.loc[len(df)] = result_A2
        result_A3 = excel_input(gameboard,fuel[i], A, 'A', h3, 'h3', i+1)
        df.loc[len(df)] = result_A3
        result_A4 = excel_input(gameboard,fuel[i], A, 'A', h4, 'h4', i+1)
        df.loc[len(df)] = result_A4
    file_name = 'algo_analysis.xlsx'
    df.to_excel(file_name)

        
```

```python
def excel_input(gameboard, fuel, algo, algo_str, h, h_str, number):
    result_list = [number, algo_str, h_str]
    if h != None:
        start = time.time()
        goalstate=algo(gameboard,fuel,h)
        end = time.time()
    else:
        start = time.time()
        goalstate=algo(gameboard,fuel)
        end = time.time()
    duration = "{:.2f}".format(end - start)
    
    tree = goalstate[0]
    search_length = goalstate[1]
    solution_length = 0
    
    if(isinstance(goalstate[0], Tree)):
        while tree.parent is not None:
            solution_length = solution_length + 1
            tree = tree.parent
    else:
        solution_length = "No solution"

    result_list.append(solution_length)
    result_list.append(search_length)
    result_list.append(duration)
    return result_list
```
